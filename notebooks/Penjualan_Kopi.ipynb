Data Penjualan

```python
import pandas as pd
import numpy as np

# Membuat data palsu untuk latihan
data = {
    'Tanggal': pd.date_range(start='2025-01-01', periods=20, freq='D'),
    'Menu': ['Kopi Susu', 'Latte', 'Espresso', 'Americano', 'Cappuccino'] * 4,
    'Harga': [20000, 25000, 15000, 20000, 25000] * 4,
    'Jumlah_Terjual': np.random.randint(10, 50, size=20),
    'Rating_Pelanggan': np.random.uniform(3.5, 5.0, size=20).round(1)
}

df = pd.DataFrame(data)
df['Total_Pendapatan'] = df['Harga'] * df['Jumlah_Terjual']

print("Data Berhasil Dibuat!")
df.head() # Menampilkan 5 data terata
```

Menu Paling Laris

```python
# Menghitung total terjual per menu
best_seller = df.groupby('Menu')['Jumlah_Terjual'].sum().sort_values(ascending=False)
print("Total Terjual Per Menu:")
print(best_seller)
```

Pendapatan Tertinggi

```python
# Menghitung total pendapatan per menu
revenue = df.groupby('Menu')['Total_Pendapatan'].sum().sort_values(ascending=False)
print("\nPendapatan Per Menu:")
print(revenue)
```

Membuat Visualisasi

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set tema visualisasi
sns.set_theme(style="whitegrid")

# Bikin Grafik Batang untuk Penjualan
plt.figure(figsize=(10, 6))
sns.barplot(x=best_seller.index, y=best_seller.values, palette="viridis")
plt.title('Menu Kopi Paling Laris (Januari 2025)', fontsize=15)
plt.xlabel('Nama Menu')
plt.ylabel('Total Cup Terjual')
plt.show()
```

Mencari Masalah (Korelasi Rating)

```python
# Cek korelasi sederhana
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Harga', y='Rating_Pelanggan', hue='Menu', s=100)
plt.title('Hubungan Harga vs Rating Pelanggan')
plt.show()
```

### Kesimpulan Analisis

**Menu Terlaris:** Menu Americano adalah yang paling diminati pelanggan dan memberikan keuntungan yang paling besar.
